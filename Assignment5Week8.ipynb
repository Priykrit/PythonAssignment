## Que  2

In Bokeh, glyphs are visual markers used to represent data points on a plot. They include shapes, lines, and other graphical elements that help visualize data. Glyphs are added to a Bokeh plot to display data in a visually meaningful way.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

p = figure(title="Glyph Example", x_axis_label="X-axis", y_axis_label="Y-axis")

p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=15, color="blue", legend_label="Circles")

p.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2, color="green", legend_label="Line")

p.title.text_font_size = "20pt"
p.legend.title = "Glyphs"
p.legend.label_text_font_size = "12pt"

output_notebook()
show(p)
```

A Bokeh server is a Python library and framework that allows you to create interactive web applications with dynamic, real-time updates. It enables you to build interactive plots, dashboards, and data-driven applications that can respond to user inputs or external data changes. Bokeh server runs a Python server in the background, enabling the application to maintain a live connection with the web browser and update the content dynamically.

## Que  5

Both Flask and Django are popular web frameworks in Python that allow you to create web applications.

from flask import Flask, render_template

from bokeh.plotting import figure, show

from bokeh.embed import components


app = Flask(__name__)

@app.route('/')

def index():

    p = figure(title="Bokeh Plot Example", x_axis_label="X-axis", y_axis_label="Y-axis")
    
    p.circle([1, 2, 3], [4, 5, 6], size=15, color="blue")

    script, div = components(p)
    
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':

    app.run()
