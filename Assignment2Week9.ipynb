## Que 5

## DataFrame.size:
* This attribute returns the total number of elements in the DataFrame.
* It calculates the size by multiplying the number of rows by the number of columns.
## DataFrame.shape:
* This attribute returns a tuple representing the dimensions of the DataFrame.
* The tuple contains two values: the number of rows and the number of columns.

```python
data = {'A':[1,2,3,8,7],'Values':[4,5,6,43,2],'C':[7,8,9,67,21]}
df = pd.DataFrame(data)
print("size",df.size)
print("shape",df.shape)
```
